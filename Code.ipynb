**1. Tesseract OCR (tesseract-ocr)**

Tesseract OCR is an optical character recognition (OCR) engine. It's a command-line tool that can recognize text within images and convert it into editable text.

**2. Pytesseract (pytesseract)**

Pytesseract is a Python wrapper for Tesseract OCR. It allows you to use Tesseract OCR functionality within your Python code. You can use it to extract text from images directly in your Python scripts.

**3. gTTS (gTTS)**

gTTS (Google Text-to-Speech) is a Python library that interfaces with Google Translate's text-to-speech API. It enables you to convert text into spoken audio. You can generate audio files in various languages and accents.

**4. Pillow (Pillow==9.0.0)**

Pillow is a powerful image processing library in Python. It's a fork of the PIL (Python Imaging Library) and provides extensive capabilities for working with images. You can use Pillow to open, manipulate, and save images in various formats. In this case, it's likely used to load and preprocess images before feeding them to Tesseract OCR.

```python
import pytesseract
from PIL import Image
import gtts
```

```python
img = Image.open('/content/Picture1 (1).png')  # Using Image function to take image input
print(img)       # Printing image size and memory location
```

```python
result = pytesseract.image_to_string(img)     # Extracting the text from img and storing it in result
print(result)       # Printing the extracted text to console
tts = gtts.gTTS(result)   # Takes extracted text and creates a text to speech object
tts.save("hello.mp3")     # Saves the audio file in mp3 format
```
